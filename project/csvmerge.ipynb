```python
import os
import pandas as pd
```

```python
total = []
for file in os.listdir('data/total'):
    path = os.path.join('data/total',file)
    with open(path,'r',encoding='utf-8') as f:
        df = pd.read_csv(path)
        year = os.path.splitext(file)[0]
        df['year'] = int(year)
        total.append(df)
    print(file)
    print(df.dtypes)
ret = pd.concat(total,ignore_index=True)
```

```python
ret.to_csv('data/total.csv',index=False,encoding='utf-8')
```
