```python
import json
import os
```

```python
total = []
for file in os.listdir('geojson'):
    path = os.path.join('geojson',file)
    with open(path,'r',encoding='utf-8') as f:
        geo = json.load(f)
        features = geo['features']
        total += features
    print(file)
ret = { "type": "FeatureCollection","features":total}
with open('total.json','w',encoding='utf-8') as f:
    json.dump(ret,f,ensure_ascii=False)
```
